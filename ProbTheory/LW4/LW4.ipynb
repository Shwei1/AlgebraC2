## 1 Від'ємний біноміальний розподіл

### 3 Розподіл Пуассона

Зобразіть графічно функцію щільності, функцію розподілу та гістограму $10000$ випадкових величин розподілу Пуассона з параметрами:
1. $\mu = 0.1$
2. $\mu = 0.2$
3. $\mu = 0.3$
4. $\mu = 0.4$
5. $\mu = 0.5$
6. $\mu = 0.6$
7. $\mu = 0.7$
8. $\mu = 0.8$
9. $\mu = 0.9$

```python
mus = np.arange(0.1, 0.9, 0.1)
np.random.seed(42)
mu = np.random.choice(mus)
x2 = np.arange(0, 15)
f_z = poisson.pmf(x2, mu)
F_z = poisson.cdf(x2, mu)
z = poisson.rvs(mu, size=10000)

fig, axs = plt.subplots(1, 3, figsize=(15, 4))

axs[0].stem(x2, f_z)
axs[0].set_title('PMF')
axs[0].set_xlabel('x')
axs[0].set_ylabel('P(X=x)')
axs[0].grid(True)

axs[1].step(x2, F_z, where='mid')
axs[1].set_title('CDF')
axs[1].set_xlabel('x')
axs[1].set_ylabel('P(X≤x)')
axs[1].grid(True)

axs[2].hist(z, bins=np.arange(0, 15)-0.5, edgecolor='black', alpha=0.7)
axs[2].set_title('Histogram')
axs[2].set_xlabel('x')
axs[2].set_ylabel('count')
axs[2].grid(True)

plt.tight_layout()
plt.show()

print(np.unique(z, return_counts=True))



```

## 4 Апроксимація біноміального розподілу розподілом Пуассона

Покажіть схожість біноміального розподілу та розподілу Пуассона (їх функцій щільності) з відповідними параметрами.
Наприклад:
- $n = 100, p = 0.05$
- $n = 100, p = 0.1$
- $n = 100, p = 0.15$
- $n = 100, p = 0.2$
- $n = 100, p = 0.35$
- $n = 100, p = 0.5$
- $n = 100, p = 0.75$

```python
n = 100
ps = np.array([0.05, 0.1, 0.15, 0.2, 0.35, 0.5, 0.75])
p = np.random.choice(ps)
mu = n * (1 - p) / p

x = np.arange(int(mu - 5 * np.sqrt(mu)), int(mu + 5 * np.sqrt(mu)))

nb_pmf = nbinom.pmf(x, n, p)
poisson_pmf = poisson.pmf(x, mu)

plt.figure(figsize=(10, 5))
plt.plot(x, nb_pmf, label='NBinomial', color='blue')
plt.plot(x, poisson_pmf, label='Poisson', color='red', linestyle='--')
plt.legend()
plt.title("NBinomial vs. Poisson")
plt.xlabel('x')
plt.ylabel('Probability')
plt.grid(True)
plt.show()

```
