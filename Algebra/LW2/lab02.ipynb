## Завдання 2. 
__Для кожної з визначених дій перевірити чи буде дія асоціативною. Якщо дія не асоціативна, то вивести трійку, для якої ця умова порушується.__

```python
def is_associative(space, action): 
    for a in space:
        for b in space:
            for c in space:
                if action((action((a, b)), c)) != action((a, action((b, c)))):
                    counter = (a, b, c)
                    return False, counter
    return True, None

M = FiniteSetMaps(product(needed_set, repeat=2), needed_set)

# needed_set = [1, 2]
needed_set = [1, 2, 3]

associative_actions = []
with open("output_2_2.txt", "w") as file:
    print(' ', file=file)
    for f in all_actions(needed_set): 
        result, counter = is_associative(needed_set, f)
        if not result: 
            print("Дія не асоціативна, контрприклад:", counter, file=file)
        else: 
            associative_actions.append(f)
with open("output_2_2.txt", "r+") as file:
    file.seek(0)
    print(f'К-ть асоціативних дій для {len(needed_set)}-елементної множини:', len(associative_actions), file=file)
```

## Завдання 3. 

__Перевірити, які з визначених множин з асоціативними бінарними діями будуть ізоморфними між собою і знайти кількість класів попарно ізоморфних між собою множин.__

```python
def is_isomorphic(space1, action1, space2, action2): 
    for phi in permutations(space1):
        phi_map = {a: phi[i] for i, a in enumerate(space1)}
        is_iso = True
        for a in space1:
            for b in space1:
                if phi_map[action1((a, b))] != action2((phi_map[a], phi_map[b])):
                    is_iso = False
                    break
            if not is_iso:
                break
        
        if is_iso:
            return True  
    
    return False


isomorph_classes = []

for f1 in associative_actions: 
    for iso_class in isomorph_classes: 
        f2 = iso_class[0]
        if is_isomorphic(needed_set, f1, needed_set, f2): 
            iso_class[1] += 1
            break 
    # якщо жоден з представників не ізоморфний 
    else:    
        isomorph_classes.append([f1, 1])
with open("output_2_3.txt", "w") as file:
    print('К-ть класів ізоморфності:', len(isomorph_classes), file=file)
    print('К-ть дій в кожному класі:', file=file)
    for iso_class in isomorph_classes: 
        print(*iso_class, file=file)
```
