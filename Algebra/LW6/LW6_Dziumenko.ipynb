## Завдання 1

В групі рухів правильного восьмикутника задайте циклічну підгрупу порядку 4. Побудуйте для неї ліві і праві класи суміжності та перевірте чи буде ця група нормальною. 

```python
D8 = DihedralGroup(8)
Z4 = D8.subgroup(['(1,3,5,7)(2,4,6,8)']) 
Z4.order()
right_cosets = D8.cosets(Z4, side='right')
left_cosets = D8.cosets(Z4, side='left')
print("Right cosets of D8 by Z4: ", right_cosets)
print("Left cosets of D8 by Z4: ", left_cosets)

sorted_right_cosets = sorted([sorted(coset) for coset in right_cosets])
sorted_left_cosets = sorted([sorted(coset) for coset in left_cosets])
condition_for_normality = sorted_right_cosets == sorted_left_cosets
print("Is Z4 a normal subset of D8: ", condition_for_normality)
```

## Завдання 2 

Переберіть всі підгрупи знакозмінної групи $A_5$. Перевірте всі на нормальність. Порівняйте час роботи з методом `.is_simple` для групи $A_5$. 

```python
import time

def check_normality(group, subgroup):
    subgroup = group.subgroup([str(subgroup)])
    right_cosets = group.cosets(subgroup, side='right')
    left_cosets = group.cosets(subgroup, side='left')
    sorted_right_cosets = sorted([sorted(coset) for coset in right_cosets])
    sorted_left_cosets = sorted([sorted(coset) for coset in left_cosets])
    return sorted_right_cosets == sorted_left_cosets

A5 = AlternatingGroup(5)
with open("output_6_2.txt", 'w') as f:
    begin_time = time.time()
    normal_subgroups = [check_normality(A5, subgroup) for subgroup in A5]
    print("Conclusion about simplicity of A5: ", (lambda subgroup: normal_subgroups.count(True) == 1)(subgroup), file = f)
    end_time = time.time()
    for subgroup in A5:
        print(f"Is the subgroup generated by {subgroup} normal in A5:", check_normality(A5, subgroup), file=f)
    print("This method by direct comparison took", end_time-begin_time, " seconds to complete.", file=f)
    begin_time = time.time()
    print(f"is A5 a simple group:", A5.is_simple(), file=f)
    end_time = time.time()
    print("This method by the built-in function took", end_time-begin_time, " seconds to complete.", file=f)
    
```

Задайте їх як елементи групи $S_{100}$ і породіть ними підгрупу $G$. Це так звана Higman-Sims група. Виведіть її порядок і знайдіть нормальні підгрупи. Зробіть висновки. 

```python
with open('HSG1-p100B0.g1', 'r') as file1:
    generator1 = file1.read().strip()
    generator1 = generator1.replace('a11 :=', '').replace(';', '').replace('\n', '').replace(' ', '').strip()

with open('HSG1-p100B0.g2', 'r') as file2:
    generator2 = file2.read().strip()
    generator2 = generator2.replace('a21 :=', '').replace(';', '').replace('\n', '').replace(' ', '').strip()

S100 = SymmetricGroup(100)
HGS = S100.subgroup([generator1, generator2])

print("The order of the Higman-Sims group is", HGS.order())

print("Subgroups of the Higman-Sims group are\n")
for i in HGS.normal_subgroups(): #Тільки тривіальні нормальні підгрупи, отже ця група — проста.
    print(i, end='\n\n')

print("Is the Higman-Sims group simple:", HGS.is_simple())

# print(generator1)
# generator2
```
