## Завдання 1. 

__Знайдіть усі незвідні системи твірних групи $S_4$.__

```python
S4 = SymmetricGroup(4)

generators = set()
# min_generators = []

for i in S4:
    for j in S4:
        if S4.order() == S4.subgroup([i, j]).order():
            generators.add(tuple(sorted([i, j]))) # Якщо пара елементів породжує групу S_n, то ця твірна — незвідна.
                                      # Як і правильно те, що будь-яка твірна S_n з 3+ елементами буде звідною. 
with open("output_4_1.txt", 'w') as f:
    for gen in generators:
        print(f"Generator of S4 consisting of elements {gen}.\n", file=f)




# for generator in generators: #Тут про всяк можна перевірити, що одним елементом S4 не породиться
#     is_min = True
#     for el in generator:
#         smaller_generator = [x for x in generator if x != el]
#         if S4.order() != S4.subgroup(smaller_generator).order():
#             is_min = False
#     if !is_min:
#         min_generators.append(generator)
        

# min_generators

# for i in S4:
#     print(i, "\n")

```

## Завдання 2. 

__Знайдіть усі підгрупи груп $S_4, A_4, S_5, A_5$.__

```python
S4 = SymmetricGroup(4)
S5 = SymmetricGroup(5)
A4 = AlternatingGroup(4)
A5 = AlternatingGroup(5)

with open("output_4_2.txt", 'w') as f:
    print("Subgroups of S4: ", file=f)
    for i in S4.subgroups():
        print(i, file=f)
    print('\n', file=f)
    print("Subgroups of S5: ", file=f)
    for i in S5.subgroups():
        print(i, file=f)
    print('\n', file=f)
    print("Subgroups of A4: ", file=f)
    for i in A4.subgroups():
        print(i, file=f)
    print('\n', file=f)
    print("Subgroups of A5: ", file=f)
    for i in A5.subgroups():
        print(i, file=f)
    print('\n', file=f)
    
```

## Завдання 3. 

__Задайте дві випадкові підстановки з групи $S_n, \, (1 \le n \le 100)$. Перевірте чи є вони парними. Знайдіть порядок групи, ними породженої. Чи буде ця група дорівнювати $A_n$? $S_n$?__

```python
import random

n = int(input())
Sn = SymmetricGroup(n)
An = AlternatingGroup(n)
k = randint(1, n)

def rand_permutation(n, k):
    return tuple(random.sample(range(1, n+1), k))

perm1 = rand_permutation(n, k)
perm2 = rand_permutation(n, k)

print(f"Permutation {perm1} is even") if sign(Sn([perm1])) == 1 else print(f"Permutation {perm1} is odd")
print(f"Permutation {perm2} is even") if sign(Sn([perm2])) == 1 else print(f"Permutation {perm2} is odd")

G = Sn.subgroup([perm1, perm2])

print(f"Order of group, generated by {perm1} and {perm2} is {G.order()}")
print(f"This group is equivalent to S{n}") if Sn.order() == G.order() else print(f"This group is not equivalent to S{n}")
print(f"This group is equivalent to A{n}") if An.order() == G.order() else print(f"This group is not equivalent to A{n}")




    





```

```python
n = int(input())

An = AlternatingGroup(n)
# An = GeneralLinearGroup(2, n)

ordnung = An.cardinality()
div = divisors(ordnung)
subgroup_orders = [int(j.order()) for j in An.subgroups()]

Lagrange_complete = True
for i in div:
    if i not in subgroup_orders:
        Lagrange_complete = False
        print(f"A{n} does not have a subgroup of order {i}")

if Lagrange_complete:
    print(f"A{n} has subgroups of every order from the set of A{n} order's divisors")


k = int(input())
PSL2k = PSL(2, k)

ordnung = PSL2k.cardinality()
div = divisors(ordnung)
subgroup_orders = [int(j.order()) for j in PSL2k.subgroups()]

Lagrange_complete = True
for i in div:
    if i not in subgroup_orders:
        Lagrange_complete = False
        print(f"PSL_2_{k} does not have a subgroup of order {i}")

if Lagrange_complete:
    print(f"PSL_2_{k} has subgroups of every order from the set of PSL_2_{k} order's divisors")









```
