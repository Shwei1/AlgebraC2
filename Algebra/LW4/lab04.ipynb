## Завдання 2.

## Завдання 3. 

__Задайте дві випадкові підстановки з групи $S_n, \, (1 \le n \le 100)$. Перевірте чи є вони парними. Знайдіть порядок групи, ними породженої. Чи буде ця група дорівнювати $A_n$? $S_n$?__

```python
import random

n = int(input())
Sn = SymmetricGroup(n)
An = AlternatingGroup(n)
k = randint(1, n)

def rand_permutation(n, k):
    return tuple(random.sample(range(1, n+1), k))

perm1 = rand_permutation(n, k)
perm2 = rand_permutation(n, k)

print(f"Permutation {perm1} is even") if sign(Sn([perm1])) == 1 else print(f"Permutation {perm1} is odd")
print(f"Permutation {perm2} is even") if sign(Sn([perm2])) == 1 else print(f"Permutation {perm2} is odd")

G = Sn.subgroup([perm1, perm2])

print(f"Order of group, generated by {perm1} and {perm2} is {G.order()}")
print(f"This group is equivalent to S{n}") if Sn.order() == G.order() else print(f"This group is not equivalent to S{n}")
print(f"This group is equivalent to A{n}") if An.order() == G.order() else print(f"This group is not equivalent to A{n}")




    





```
